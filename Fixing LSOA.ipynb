```python
import pandas as pd
import geopandas as gpd
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

# LSOA

```python
EW_lsoa = gpd.read_file("Data/England_lsoa_2021/england_lsoa_2021.shp")
manchester_lsoa = gpd.read_file("Data/manchester_lsoa/manchester_lsoa.shp")
```

```python
EW_lsoa.head()
```

```python
missing_row = EW_lsoa[EW_lsoa['lsoa21cd'] == "E01005866"]
```

```python
# Check if the row exists
if not missing_row.empty:
    # Concatenate the missing row to the Greater Manchester GeoDataFrame
    manchester_lsoa = pd.concat([manchester_lsoa, missing_row], ignore_index=True)
else:
    print(f"LSOA code E01005866 not found in the England and Wales GeoDataFrame.")
```

```python
manchester_lsoa.plot()
```

```python
# Save the updated Greater Manchester GeoDataFrame if needed
manchester_lsoa.to_file('Data/manchester_lsoa/manchester_lsoa.shp')
```
